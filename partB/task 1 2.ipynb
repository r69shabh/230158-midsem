# Task 1.2: Key Assumptions

## Assumption 1
**Assumption:** The current online classifier is reliable enough to decide when a badly misclassified point is probably not worth updating on.  
**Why the method needs it:** LASVM-I uses the current margin value to filter examples before they reach the usual LASVM update. If the current separator is too unstable, the algorithm can throw away useful hard points just because they happen to fall below the left flat region early in training.  
**Violation scenario:** A dataset with strong distribution shift or a very poor early initialization could make many valid boundary points look like extreme outliers. In that case, the filtering rule would block learning instead of protecting it.  
**Paper reference:** Section 6 (LASVM-I filtering rule), linked back to the Ramp-loss shape introduced in Eq. (14).

## Assumption 2
**Assumption:** Some severely misclassified examples are genuinely noisy, mislabeled, or outlying, so ignoring them can improve the classifier.  
**Why the method needs it:** The whole point of moving from hinge loss to Ramp loss is that not every mistake should keep increasing the penalty. If the data do not actually contain harmful noise or outliers, the extra left-side filtering gives away information without much benefit.  
**Violation scenario:** On a clean dataset where almost every hard point is informative, the nonconvex filter may remove rare but useful support vectors. Then the standard convex SVM can match or beat it.  
**Paper reference:** Section 5 for the Ramp-loss motivation and Section 8 for the noisy-data experiments.

## Assumption 3
**Assumption:** The most informative training points are concentrated near the margin, not far on either flat side.  
**Why the method needs it:** LASVM-I only spends computation on points with `s \le z \le 1`, which means it is betting that the action is near the margin. This is what lets the method behave like a selective online learner instead of treating every sample equally.  
**Violation scenario:** If the useful signal comes from far-misclassified minority points or from examples that only become useful after several passes, the method can become too conservative. It will then underfit because the very examples it should eventually learn from are skipped too early.  
**Paper reference:** Section 6 for the LASVM-I rule and the discussion around the active-learning interpretation when `s = -1`.
