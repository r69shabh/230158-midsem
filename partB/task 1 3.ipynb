# Task 1.3: What the Paper Claims to Improve

The main baseline in the paper is the standard **convex hinge-loss online SVM / LASVM** family. In practical terms, the comparison is against the usual online large-margin learner that keeps updating on every point with margin below 1.

The limitation the paper points out is that hinge loss keeps penalizing points that are very far on the wrong side of the boundary. That means a noisy label or a strong outlier can still become influential and even end up as a support vector, which is exactly the opposite of what you want when the data stream is messy.

The proposed method tries to fix this by replacing hinge loss with Ramp loss and then building LASVM-NC / LASVM-I on top of that idea. In LASVM-I, the important operational change is simple: if a point is too far below the `s` threshold, it is ignored instead of being allowed to keep dragging the separator.

One situation where I would not expect the paper's method to outperform the convex baseline is a mostly clean dataset with only mild label noise and several genuinely difficult points near class overlap. In that case, the left-side filter can discard examples that are hard but still useful, so the learner may trade away a little accuracy just to save updates. I would also expect failure when `s` is set too aggressively high, because then the model starts ignoring even within-margin points that should have become support vectors.
